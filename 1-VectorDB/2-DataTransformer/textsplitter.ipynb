```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string="""

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Simple HTML Structure</title>
</head>
<body>
    <div>
        <h1>Welcome to My Webpage</h1>
        <br>
        <h2>Introduction</h2>
        <div>
            <h3>What is this page about?</h3>
            <br>
            <div>This is a simple HTML page designed to demonstrate the use of specific tags like <strong>h1</strong>, <strong>h2</strong>, <strong>h3</strong>, <strong>body</strong>, <strong>br</strong>, and <strong>div</strong>.</div>
        </div>
        <br>
        <h2>Content Section</h2>
        <div>
            <h3>Why are these tags important?</h3>
            <br>
            <div>
                These tags are essential for structuring content on the web:
                <br>
                <div>
                    - `<h1>`: Defines the main heading of the page.
                    <br>
                    - `<h2>`: Defines subheadings.
                    <br>
                    - `<h3>`: Provides further sub-sections under `<h2>`.
                    <br>
                    - `<div>`: Groups content into sections.
                    <br>
                    - `<br>`: Adds line breaks for formatting.
                    <br>
                </div>
            </div>
        </div>
    </div>
</body>
</html>

"""
```

```python
headers_to_split_on = [
    ("h1","Header 1"),
    ("h2","Header2"),
    ("h3","Header3")
]
```

```python
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
```

```python
html_header_splits
```

```python
url = "https://plato.stanford.edu/entries/plato/"

headers_to_split_on = [
    ("h1","Header 1"),
    ("h2","Header2"),
    ("h3","Header3"),
    ("h4","Header4")
]
```
